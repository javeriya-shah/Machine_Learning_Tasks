##Analysing and finding Relationships in the given Dataset

```
plt.figure(figsize=(60,60))
sns.barplot(x='region_txt' , y='region' , data=dataset)
```

The above plot is a bar graph plotted between Region and Region_txt and it shows all the regions coressponding to the number assigned to them in the database.


```
plt.figure(figsize=(20,20))
sns.barplot(x='attacktype1_txt' , y='attacktype1' , data=dataset)
```

The above plot is a bar graph plotted between Attacktype1 and Attacktype1_txt and it shows all the Attacktypes coressponding to the number assigned to them in the database.


```
plt.figure(figsize=(60,60))
plt.axis([0,15, 1950,2000])
sns.barplot(x='region' , y='iyear' ,hue='attacktype1_txt' , data=dataset)
```

The above bar graph plotted between Region and Year with hue as attacktype1_txt. This will help us know the frequency of attacks in a specific region and how much are they prevelant. Helping us know the hot zones of terrorism.

```
plt.figure(figsize=(25,25))
sns.countplot(x='multiple', hue = 'region_txt' , data=dataset)
```

```
plt.figure(figsize=(25,25))
sns.countplot(x='multiple', hue = 'attacktype1_txt' , data=dataset)
```

From the above  graphs ,  following can be inferred :
1. Multiple attacks have not been reported in many regions. Regions where they have been reported are North America , Western Europe and Middle East and north Africa. North America having a higher number.

2. Out of these multiple attacks , bombing and explosion are more frequent , then come facility/infrastructure attack , then is hijacking , Armed assaults , followed by hostage taking.

3. From the plot between iyear and regions , it can be seen that the hot zone of terrorism are North America , Middle East and North Africa  ,  south America and Western Europe.
