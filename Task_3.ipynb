##Exploratory Data Analysis


##Importing the Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##The Data

```
dataset_1 = pd.read_csv('/content/SampleSuperstore.csv')
dataset_1.head(10)
```

##Removing Negative values

```
dataset = dataset_1._get_numeric_data()
```

```
dataset[dataset.Profit < 0] = 0
```

```
dataset.head()
```

##Misssing Data


```
dataset.isnull().sum().sort_values(ascending = False)
```

##Dropping a Column

```
new_dataset = dataset.drop(['Postal Code'] , axis=1)
```

```
new_dataset.head()
```

##Analysis of Relationship between variables


```
#sns.pairplot(new_dataset)
plt.figure(figsize=(50,50))
sns.barplot(x='Sub-Category' , y='Profit', data = dataset_1)
```

The above bar plots show us that machines and copiers have contributed greatly to profit .
whereas fasteners , art and supplies have very low impact on profit.

```
plt.figure(figsize=(50,50))
sns.barplot(x='State' , y='Profit', data = dataset_1)
```

Vermont region has contributed to maximum profit followed by Rhode Island , Indiana , Minnesota .

```
sns.relplot(x='Sales' , y='Profit', hue='Category' ,data = dataset_1)
```

On analysing the sample store data and from the above plot we can aconclude that people our more intented to buy Technology related products and as it's sale increases the profit also increases.
people are also attracted to office supplies and they too have contributed towards profit to some extent.
But , it can be seen that Furniture doesn't excite the customers much and it has less sales and profit.

```
sns.barplot(x='Discount' , y='Profit', hue='Category' ,data = dataset_1)
```

Discount has a moderate effect on profit , people tend to buy technology related stuff even on 0% discount but the profit is less , but with discount upto 10% on technology related stuff the profit increases. 
discounts greater than 10% have some sales for technology related stuff but the profit is low.  
The profit through other categories i.e. office supplies and Furniture is almost negligible even if the discount was increased upto 40% . 

```
sns.barplot(x='Quantity' , y='Profit', hue='Category' ,data = dataset_1)
```

The affect of quantity is not very significant on profit ,  its unorederd and timely.
